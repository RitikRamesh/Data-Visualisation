```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = ['jan','feb','mar','apr','may','jun','july','aug','sep','oct','nov','dec']
max_temp = [2,4,8,14,20,24,28,27,23,16,9,4]
min_temp = [1,2,5,9,15,17,21,22,16,3,2,1]
avg_temp = [3,5,8,12,17,24,28,23,16,8,4,2]
```

```python
plt.plot(months, max_temp, color = 'orange',linewidth=4)
plt.show()
```

```python
plt.plot(months, max_temp, color = 'orange',linewidth=1,marker='o')
plt.show()
```

```python
plt.plot(months, max_temp, color = 'orange',marker='2')
plt.show()
```

```python
plt.plot(months, max_temp, color = 'orange',linewidth=1,marker='o',label='max')
plt.xlabel('Months')
plt.ylabel('Temperatures')
plt.title('Tmperatures range over a year')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.plot(months, max_temp, color = '#e76f51',linewidth=1,marker='o',label='max')
plt.plot(months, min_temp, color = '#e9c46a',linewidth=1,marker='o',label='min')
plt.plot(months, avg_temp, color = '#2a9d8f',linewidth=1,marker='o',label='avg')
plt.xlabel('Months')
plt.ylabel('Temperatures')
plt.title('Tmperatures range over a year')
plt.legend(shadow=1)
plt.grid(alpha=0.25)
plt.show()
```

```python
plt.scatter(months,avg_temp,s=200,color='#e76f51')
plt.scatter(months,min_temp,s=200,color='#e9c46a')
plt.scatter(months,max_temp,s=200,color='#2a9d8f')
plt.show()
```

```python
width = 0.4
p = np.arange(len(months))
p1 = [j+width for j in p]
plt.bar(p,max_temp,width,color='#e9c46a',label='max')
plt.bar(p1,min_temp,width,color='#2a9d8f',label='min')
plt.xlabel('months')
plt.ylabel('temperature')
plt.title('Temp varying over a year')
plt.xticks(p+width/2,months)
plt.legend()
plt.grid()
plt.show()
```
